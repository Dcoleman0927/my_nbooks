```python
import pandas as pd
import seaborn as sns
```

```python
avocado = pd.read_csv('avocado.csv')
avocado
```

```python
avocado.head(5)
```

```python
avocado.info()
```

```python
avocado["Date"] = pd.to_datetime(avocado["Date"])
```

```python
avocado.info()
```

```python
avocado['Date'] = avocado.Date.dt.strftime("%Y")
```

```python
avocado.query('Date == 2015 and type == "conventional" and region == "Albany"')
avocado[['Date', 'Total Bags', 'Small Bags']].head(5)
```

```python
avocado_melted = pd.melt(avocado, id_vars='Date',
                         value_vars=['Total Bags', 'Small Bags'],
                         var_name ='Bags', value_name ='Count')
avocado_melted
```

```python
sns.relplot(data=avocado_melted, kind = 'line', x='Date', y='Count', hue='Bags')
```

```python
avocado['Sales Volume'] = pd.qcut(avocado['Total Volume'], 
                                 q=4, 
                                 labels=['Poor', 'Modest', 'Good', 'Excellent'])
```

```python
avocado.head(5)
```

```python
sns.countplot(avocado, x='year', hue = 'Sales Volume')
```

```python
avocado_melted = pd.melt(avocado, id_vars='Date',
                         value_vars=['Small Bags', 'Large Bags', 'XLarge Bags'],
                         var_name ='Bags', value_name ='Count')
avocado_melted
```

```python
sns.relplot(data=avocado_melted, kind='line', x='Date', y='Count', hue='Bags')
```
